```python
# Step 2:  using a list of artists from the Kaggle dataset, 
#          gets each individual artist's Spotify ID (which will be used for iteration later)

start_time = datetime.datetime.now()

d = []

for artist in artist_list:
    
    results = sp.search(q=f'artist:{artist}', limit=1)

    
    for idx, track in enumerate(results['tracks']['items']):
        d.append (
            {
                'Artist_ID' : track['artists'][0]['id'],
                'Artist' : track['artists'][0]['name']
            }
        )

df = pd.DataFrame(d)

end_time = datetime.datetime.now()

elapsed_time = end_time - start_time
elapsed_time
```

```python
# Step 3:  creates a list of artist IDs (used in Step 4)
artist_id_list = df['Artist_ID'].tolist()
```

```python
# Step 4:  using the artist IDs from step 3, gets all the albums for each artist

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id='client_id',
                                                           client_secret='client_secret'))

start_time = datetime.datetime.now()

albums = []

for artist_id in artist_id_list:

    results = sp.artist_albums(artist_id, album_type='album')
    albums.extend(results['items'])
    while results['next']:
        results = sp.next(results)
        albums.extend(results['items'])
    seen = set()  # to avoid duplicates
    albums.sort(key=lambda album: album['name'].lower())
    for album in albums:
        name = album['name']
        if name not in seen:
            seen.add(name)

end_time = datetime.datetime.now()

elapsed_time = end_time - start_time
elapsed_time
```

```python
# Step 5:  using the 'albums' object from Step #4, builds a clean dataframe of albums for each artist
#          (includes the associated album ID, which is used in Step #6)

start_time = datetime.datetime.now()

album_list = []

for idx, album in enumerate(albums):
    
    try:
    
        album_list.append (
            {
                'Artist_ID' : album['artists'][0]['id'],
                'Artist_Name' : album['artists'][0]['name'],
                'Album_ID' : album['id'],
                'Album_Name' : album['name'],
                'Total_Tracks' : album['total_tracks'],
                'Release_Date' : album['release_date'],            
                'Spotify_URL' : album['external_urls']['spotify'],
                'Album_Image' : album['images'][1]['url']

            }
        )
    
    except IndexError:

        album_list.append (
            {
                'Artist_ID' : album['artists'][0]['id'],
                'Artist_Name' : album['artists'][0]['name'],
                'Album_ID' : album['id'],
                'Album_Name' : album['name'],
                'Total_Tracks' : album['total_tracks'],
                'Release_Date' : album['release_date'],            
                'Spotify_URL' : album['external_urls']['spotify'],
                'Album_Image' : ''

            }
        )
    

df_albums = pd.DataFrame(album_list)

end_time = datetime.datetime.now()

elapsed_time = end_time - start_time
elapsed_time
```

```python
# Step 6:  using the data from Step #5, creates a list of album IDs for all artists
album_id_list = df_albums['Album_ID'].tolist()
```

```python
# Step 7:  using the list of album IDs from Step #6, gets all of the tracks on a given album

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id='client_id',
                                                           client_secret='client_secret'))

start_time = datetime.datetime.now()

tracks = []

for album_id in album_id_list:

    results = sp.album_tracks(album_id)
    tracks.extend(results['items'])
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    seen = set()  # to avoid duplicates
    tracks.sort(key=lambda album: album['name'].lower())
    for album in tracks:
        name = album['name']
        if name not in seen:
            seen.add(name)

end_time = datetime.datetime.now()

elapsed_time = end_time - start_time
elapsed_time
```
